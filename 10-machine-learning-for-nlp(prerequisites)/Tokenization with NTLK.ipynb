```python
corpus = """Hello, I am Rithick. I am new to NLP.
I am excited to learn NLP's ecosystem of libraries! 
The syllabus is very intersting and comprehensive."""
```

```python
print(corpus)
```

```python
## Toeknization
## Paragraphs --> Sentences
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

punkt is an unsupervised sentence tokenizer trained on large text data.
sent_tokenize() in NLTK uses a pre-trained sentence boundary detection model called Punkt.

For example:

Does a period (.) always end a sentence?
→ No: e.g., “Dr. Smith is here.”

What about abbreviations?

What about “...”?

What about numbers like “3.14”?

What about names like “U.S.A.”?

Because of these complexities, NLTK does not hard-code rules.
Instead, it uses Punkt, which is a machine-learned model trained on large corpora to identify sentence boundaries.

```python
from nltk.tokenize import sent_tokenize
```

```python
documents = sent_tokenize(corpus)
```

```python
documents
```

```python
## Tokenization
## Paragraphs --> Words
## Sentences --> Words
from nltk.tokenize import word_tokenize
```

```python
words = word_tokenize(corpus)
```

```python
words
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize # tokenizes punctuations as well
```

```python
wordpunct_tokenize(corpus)
```
